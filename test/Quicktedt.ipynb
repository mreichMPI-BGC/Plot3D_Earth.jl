```julia
using GLMakie
x = range(0,10,30)
y = range(0,10,30)
z =Observable([sin(xi) + cos(yi) for xi in x, yi in y])
fig = surface(z)
```

```julia
zorig=z[]
anim = Record(fig, 1:100; framerate = 30) do zz
    z[] = zorig .* zz/50
    
end
anim
```

```julia
using Base64
function display_mp4(filename)
    display("text/html", string("""<video autoplay controls><source src="data:video/x-m4v;base64,""",
    Base64.base64encode(open(read,filename)),"""" type="video/mp4"></video>"""))
end
display_mp4("./output_NEEdiurnal.mp4")
```
